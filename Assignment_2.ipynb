# **6. Which Classroom is less busy in overall week.**

```python
from collections import Counter

room_usage = Counter()

for df in con:
    for time_index in df.index:
        if 'Room' in df.columns:
            room = df.loc[time_index, 'Room']
            if room != 'Free Slot':  # Only count actual usage
                room_usage[room] += 1

# Get the least busy classroom
if room_usage:
    least_busy_classroom = room_usage.most_common()[-1]
    print(f"Least busy classroom: {least_busy_classroom[0]} (Used {least_busy_classroom[1]} times)")
else:
    print("No classroom usage found.")
```

# **7. Which Lab is the busiest lab in the overall week, return its name.**

```python
from collections import Counter

lab_usage = Counter()

for df in con:
    for time_index in df.index:
        if 'Room' in df.columns:
            room = df.loc[time_index, 'Room']
            if isinstance(room, str) and 'lab' in room.lower():  # Check if it's a lab
                if room.lower() != 'free slot':
                    lab_usage[room] += 1

# Get the busiest lab
if lab_usage:
    busiest_lab = lab_usage.most_common(1)[0]
    print(f"Busiest Lab: {busiest_lab[0]} (Used {busiest_lab[1]} times)")
else:
    print("No lab usage found.")
```
